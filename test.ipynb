```python
import numpy as np
from PIL import Image
import os

# サンプルデータを確認
sample_dir = "s1/00202"
image_file = os.path.join(sample_dir, "001.bmp")
landmark_file = os.path.join(sample_dir, "001_landmarks68.npy")

# 画像を読み込み
img = Image.open(image_file)
print(f"画像サイズ: {img.size}")
print(f"画像モード: {img.mode}")

# ランドマークデータを読み込み
landmarks = np.load(landmark_file)
print(f"ランドマークデータ形状: {landmarks.shape}")
print(f"ランドマークデータの一部:\n{landmarks[:5]}")  # 最初の5点を表示

# 画像とランドマークの可視化（オプション）
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.imshow(img, cmap='gray')
ax.scatter(landmarks[:, 0], landmarks[:, 1], c='red', s=10)
ax.set_title("サンプル画像とランドマーク")
plt.show()
```
